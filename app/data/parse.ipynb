```python
import pandas as pd
```

```python
voteview = pd.read_csv('HSall_members.csv')
voteview = voteview[voteview['congress'] > 118-25]
voteview
```

```python
party_map = {
    100: 'Democrat',
    200: 'Republican',
    112: 'Conservative',
    328: 'Independent'
}

voteview['party'] = voteview['party_code'].map(party_map)
voteview['party'].unique()
```

```python
dems = voteview[voteview['party'] == 'Democrat']
reps = voteview[voteview['party'] == 'Republican']

dems
```

```python
pd.concat([dems, reps]).to_json('subjects.json', orient='records')
```
